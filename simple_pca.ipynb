```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot, transforms
```

```python
# Given data set of 14 students' math and reading test scores
math = np.array([99,84,44,76,94,88,79,82,97,63,60,75,60,80,95,70,58,92,20,100])
reading = np.array([92,95,67,90,71,85,72,92,64,89,40,75,97,60,94,65,90,70,32,100])

# Plot Reading vs Math test scores
plt.scatter(math,reading)
plt.ylim(0,100)
plt.xlim(0,100)
plt.title('Student Reading vs Math Test Scores')
plt.xlabel('Math')
plt.ylabel('Reading')
plt.grid()
plt.savefig('simple_pca_plot1.png')
plt.show()
```

```python
# get average of test scores
math_avg = math.sum()/len(math)
reading_avg = reading.sum()/len(reading)
```

```python
# graph new plot with origin as averages obtained above
plt.scatter(math - math_avg,reading - reading_avg)

plt.axhline(y=0, lw=1, color='k')
plt.axvline(x=0, lw=1, color='k')

plt.title('Shifted Student Reading vs Math Test Scores')
plt.xlabel('Math')
plt.ylabel('Reading')
#plt.grid()
plt.savefig('simple_pca_plot2.png')
plt.show()
```

```python
# get and plot best linear fit
m, b = np.polyfit(math - math_avg,reading - reading_avg, 1)
plt.plot(math - math_avg,reading - reading_avg, 'o')
plt.plot(math - math_avg, m*(math - math_avg)+b, 'r', label='PC 1')
plt.plot(math - math_avg, -(1/m)*(math - math_avg)+b, 'b', label='PC 2')
plt.axhline(y=0, lw=1, color='k')
plt.axvline(x=0, lw=1, color='k')
plt.title('Shifted Student Reading vs Math Test Scores')
plt.xlabel('Math')
plt.ylabel('Reading')
plt.legend()
plt.xlim(-60,30)
plt.ylim(-50,30)
#plt.savefig('simple_pca_plot3.png')
plt.show()
```

```python
# rotate clockwise around origin by angle = slope
final_math = np.array([])
final_reading = np.array([])
a = np.cos(-0.442456333298)
b = -np.sin(-0.442456333298)
c = np.sin(-0.442456333298)
d = np.cos(-0.442456333298)
for i in range(len(math)):
    final_math = np.append(final_math,(math - math_avg)[i]*a + (reading - reading_avg)[i]*b)
    final_reading = np.append(final_reading,(math - math_avg)[i]*c + (reading - reading_avg)[i]*d)
```

```python
plt.scatter(final_math,final_reading)
plt.grid()
plt.axhline(y=0, lw=1, color='r', label='PC 1')
plt.axvline(x=0, lw=1, color='b', label='PC 2')
plt.title('Final PCA Plot of Shifted Student Reading vs Math Test Scores')
plt.xlabel('Math')
plt.ylabel('Reading')
plt.legend()
plt.savefig('simple_pca_plot4.png')
plt.show()
```
